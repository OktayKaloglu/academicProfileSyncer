```python

import pandas as pd
from jsonWrite import writeJson    

def print_full(x):
    pd.set_option('display.max_rows', len(x))
    print(x)
    pd.reset_option('display.max_rows')

def split_instructors(x):
    
    for sep in [',','\n']:
        if sep in x:
            x=x.split(sep)
            repX=[]
            for item in x:
                if len(item)>0:
                    if not str.isdigit(item[0]) :
                        repX.append(item)      
            #new list could contain more than 1 instructor
            x=repX
            # if list contains just one instructor
            if len(repX)==1:
                
                x=repX[0]
    return x

def onlyTheName(x):
    
    if '.' in x :
        if  x.find('.')>len(x)/3: # if any dot present in the name like middle name (just one case)
            return x
        x=x.split('.')
        x=x[-1]
        for keyword in ["GÖREVLİSİ","ÜYESİ"]:
            index=x.find(keyword)
            if index != -1 :
                x=x[index+len(keyword)+1:]
    return x.lstrip()#or strip for right side too




```

```python
print(onlyTheName("Dr. Öğr. ÜYESİ Emine SEZER"))


```

```python
instructors=pd.read_csv("teachers.csv",sep=',',names=["instructor","lesson","url"])#for loading you have  header directly use header
instructors.dropna(axis="index",inplace=True)
#convert str the instructor column for further split operations
instructors.instructor = instructors.instructor.astype(str)


instructors["instructor"]=instructors["instructor"].apply(lambda x:split_instructors(x))# if an entry has more than 1 instructor innit
instructors=instructors.explode('instructor')#if any entry splitted, they will need to have new entry for each
instructors["instructor"]=instructors["instructor"].apply(lambda x: x.strip())#for removing white spaces


instructors.instructor = instructors.instructor.apply(lambda x: (x.replace('i','İ')).upper())#for tr UPPER
instructors["instructor"]=instructors["instructor"].apply(lambda x: onlyTheName(x))#removing the '.' ones and 'üyesi' words from the names

instructors=instructors["instructor"].unique().tolist()#filter to just the unique instructor names
```
