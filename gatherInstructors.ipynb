```python
class Proxies:
    def __init__(self):
        self.timeout=25
        self.proxies={
            'http':None,
            'https':None
        }
        self.works=None
        self.set_proxies()
        
           
    def set_proxies(self):
        self.setThem(True)
        if not self.works:
            self.setThem(False)
        
    def setThem(self,ggl):
        for i in range(10):
            try:
                self.fp=FreeProxy(google=ggl,https=False)
                proxy=self.fp.get()
                self.proxies['http']=proxy
                self.proxies['https']=proxy
                if self.check():    
                    break
            except:
                continue
                
    def check(self):
        ip = self.proxies['http'].split(':')[1][2:]
        with requests.get('http://scholar.google.com', proxies=self.proxies, timeout=self.timeout, stream=True) as r:
            if r.raw.connection.sock and r.raw.connection.sock.getpeername()[0]==ip:
                self.works=True
                return True
        self.works=False
        return False
    
    def renewProxies(self):
        self.set_proxies()
```
